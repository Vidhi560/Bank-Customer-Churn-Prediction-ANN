```python
input_data = dataset.iloc[:,:-1]
output_data = dataset.iloc[:,-1:]
# for slicing 
#input = : в†’ all rows,:-1 в†’ last column (Exited) leave.
#output = : в†’ all rows,-1: в†’ last column show (exited).
```

```python
ss = StandardScaler()
#grapical data identify on the basis of measurerment(length,Breadth)
#StandardScaler scales numerical features to a common scale, helping Machine Learning models perform better.
```

```python
input_data = dataset.drop(["RowNumber","CustomerId","Surname","Exited"],axis = 1)
input_data = pd.get_dummies(input_data,drop_first=True)
input_data = pd.DataFrame(ss.fit_transform(input_data),columns=input_data.columns)
# dataset.drop= deleting all thease 4 col from dataset
#get_dummies convert categorical value in form of (0,1)
# drop_first =True =dummy data delete(ex= if male was there then it automaticall understand that female is not there)
#fit_transform() first learns the parameters from the data and then transforms the data in a single step.
#It standardizes the data (mean = 0 and standard deviation = 1) and stores it as a DataFrame.
```

```python
from keras.layers import Dense #building neural networks.
from keras.models import Sequential #build a neural network layer by layer.
```

```python
input_data.head()
```

```python
input_data.shape 
```

```python
ann = Sequential()
```

```python
ann.add (Dense(units=6,input_dim = 11,activation ="relu")) #if your data is numerical unit start with 6,input dim=11 is columns 
ann.add (Dense(units=4,activation ="relu"))
ann.add (Dense(units=2,activation ="relu"))
ann.add (Dense(units=1,activation ="sigmoid"))  
# relu=The first three hidden layers use the ReLU activation function because it helps the neural network learn complex and non-linear patterns
#efficiently.It also speeds up training and reduces the vanishing gradient problem.
#sigmoid =The last output layer uses the Sigmoid activation function because the model is performing binary classification,
#where the output is a probability between 0 and 1, representing two possible classes (e.g., Yes/No, True/False, or 0/1).
```

```python
ann.compile(optimizer = "adam", loss = "binary_crossentropy", metrics =["accuracy"])
#ann.compile() prepares the neural network for training. The Adam optimizer updates the model's weights efficiently during backpropagation.
#Binary Crossentropy is used as the loss function because the problem is a binary classification task with a Sigmoid output layer. 
#Accuracy is used as the evaluation metric to measure the percentage of correct predictions made by the model.
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
#train_test_split = its divides the dataset into training & testing sets
```

```python
x_train,x_test,y_train,y_test =train_test_split(input_data,output_data,test_size = 0.2,random_state=42)
#train_test_split() divides the dataset into training and testing sets. x_train and y_train are used to train the model, 
#while x_test and y_test are used to evaluate its performance on unseen data.
#Here, test_size=0.2 means 20% of the data is reserved for testing, and random_state=42 ensures the split is reproducible.
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
#StandardScaler is used to scale all input features to a similar range so that the model trains faster and performs better.
#We use fit_transform() on the training data and only transform() on the test data.
```

```python
ann.add(Dense(1,activation = "sigmoid"))
ann.compile(optimizer = "adam", loss = "binary_crossentropy", metrics =["accuracy"])
#Dense(1, activation="sigmoid") creates one output neuron for binary classification.compile() prepares the model for training 
#by using Adam to update weights,Binary Crossentropy to calculate the error, and Accuracy to measure the model's performance.
```

```python
ann.fit(x_train,y_train,batch_size=80,epochs=100)
# the model trains on 80 samples at a time before updating the weights, 
#and epochs=100 means the model goes through the entire training dataset 100 times.
```

```python
from sklearn.metrics import accuracy_score
#import for checking acuracy score
```

```python
ann.predict(x_test)
```

```python
prd = ann.predict(x_test)
```

```python
prd_data = []
for i in prd:
    if i [0]>0.5:
        prd_data.append(1)
    else:
        prd_data.append(0)
#for test = The model gives probabilities between 0 and 1. This code converts those probabilities into final class labels. 
#If the predicted probability is greater than 0.5, it is classified as 1; otherwise, it is classified as 0.
```

```python
prd1 = ann.predict(x_train)
prd_data1 =[]
for i in prd1:
    if i[0]>0.5:
        prd_data1.append(1)
    else:
        prd_data1.append(0)
# for train = ann.predict(x_train) predicts the output probabilities for the training data. The loop checks each probability: 
#if it is greater than 0.5, it stores 1; otherwise, it stores 0. This converts probabilities into final class labels.
```

```python
accuracy_score(y_test, prd_data)*100
```
